```python
from scipy.stats import truncnorm, rv_histogram
import numpy
import scipy
import matplotlib.pyplot as plt
from IPython.display import Image
import pandas as pd
```

# How to control crowds?
In control theory, 

1\. you identify your plant

2\. develop a controler for this plant

This can be done

a\. analytically: define ODEs which describe the plant. Determine the unknown constants of the ODEs analytically or by using measurement data if available. Use the properties of the system of ODEs to design the controler.

b\. data-driven: generate data. Use this data to approximate your plant with a surrogate model and to design a controler.

c\. mixed: define ODEs which describe the plant. Determine the unknown constants of the ODEs using a surrogate model. Use the properties of the system of ODEs to design the controler.

In this analysis, we use the analytical approach a) to design the controller.


The question is:

## How to model corridors with standard lags in a crowd dynamics control problem?

We determine the unknown constants of the ODEs analytically and compare them to measurement data to verify them.


The scenario below contains three corridors which can be interpreted as a system of parallel plants in automatic control. We try to find a simplified model for each sub plant which can be used for the design of the controler.

We measure the system response of each sub plant for a unit step. This correponds to a source where agents are constantly generated. We measure the density of agents in the measurement areas (1) - (3).

```python
dataframe = pd.DataFrame()

df = pd.read_csv("output/plant_1/densities.csv", delimiter='\s+')
timeStep = df["timeStep"]*0.4
dataframe["plant_1"] = df["areaDensityCounting-PID5"]

df = pd.read_csv("output/plant_2/densities.csv", delimiter='\s+')
dataframe["plant_2"] = df["areaDensityCounting-PID6"]

df = pd.read_csv("output/plant_3/densities.csv", delimiter='\s+')
dataframe["plant_3"] = df["areaDensityCounting-PID7"]

for plant in ["plant_1","plant_2","plant_3"]:
    plt.plot(timeStep,dataframe[plant])
    plt.xlabel("Time in s")
    plt.ylabel("Number of agents in measurement area")
    plt.title(plant)
    plt.show()
```

We observe that the number of agents in the measurement areas is zero at the beginning. Then it starts to increase after a certain deadtime and starts to osciallate around a value of 50. 

We try to find three different ODEs which approximately describe this behavior.

For that purpose, we do not longer consider the number of agents but the capacity utilization of a corridor which is measured within the measurement area. We consider a measurement area as fully occupied if there are more than 50 agents within the area. So the probability for capacity utilization is:


```python
for plant in ["plant_1","plant_2","plant_3"]:
    capacity_max = 50
    dataframe[plant] = dataframe[plant].clip(0,capacity_max)/capacity_max
    plt.plot(timeStep,dataframe[plant])
    plt.xlabel("Time in s")
    plt.ylabel("Capacity utilization")
    plt.title(plant)
    plt.show()
```

We make use of the facts that we know the distance between source and measurement area centroid and the distribution of the free flow speed.
The centroids of the measurement areas are 90m, 70m and 50m distanced from the source.
The free flow speed follows a normal distribution.

```python
my_mean = 1.34
my_std = 0.26
myclip_a = 0.5
myclip_b = 2.2

a, b = (myclip_a - my_mean) / my_std, (myclip_b - my_mean) / my_std
r = truncnorm.rvs(a, b, loc = my_mean, scale = my_std, size=10000)
plt.hist(r, density=True, histtype='stepfilled', alpha=0.2)
plt.xlabel("Free flow speed in m/s")
plt.ylabel("Probability density function")
plt.show()
```

The arrival time of an agent simply is 
$$ t = \frac{s}{v} $$

Since $v$ follows a normal distribution, the arrival time follows a reciprocal normal distribution, which we determine numerically.


```python
sample_dists = list()
for length, plant in zip([90,70,50],["plant_1","plant_2","plant_3"]):
    
    plt.hist(length/r, density=True, histtype='stepfilled', alpha=0.2, bins =50)
    sample_hist = numpy.histogram(length/r,density=True,bins=50)
    sample_dist = scipy.stats.rv_histogram(sample_hist)
    x = numpy.linspace(0,250,1000)
    plt.plot(x, sample_dist.pdf(x), color="green")
    plt.xlabel("Arrival time of agents in s")
    plt.ylabel("Probability density function d")
    plt.show()
    sample_dists.append(sample_dist)
```

In the next step, we compute the distribution functions of $d_i$. We make use of the fact that  capacity utilization is related to the arrival times. Fast agents arrive at an early point of time. Slow agents arrive at later times. We assume that the capacity utilization is maximal when all agents have arrived. If faster agents have already left the measurement area, new agents come after and replace them. This is why the distribution function of arrival times is similar to the capacity utilization. 

The green curves in the plots below depict the capacity utilization if the number of agents is infinite and new agents come after continously. The black curves is what we actually measured and serves as reference. The discrepancy between the curves is caused by multiple aspects. First, the number of agents is not infinite in reality. Second, the measure depends on the definition of the measurement area. Third, when agents leave the measurement area, new agents do not come after instantly. This is why the curves differ.

In the next step we try to find an ODE for which the solution is similar to the measurement curve. Since we know that there is a delay in the system we suggest to use a delay lag for modeling. A delay lag of n-th order corresponsd to a linear ODE of order n. Since the beginning is smooth, we suggest to use 

$$ G(s) = \frac{1}{(Ts+1)^n} $$


which is the most simple lag with smooth increase.  

We determine the time constant $T$ and find that it is somehow related to the standard deviation of the arrival time. 


```python
for length, plant, sample_dist in zip([90,70,50],["plant_1","plant_2","plant_3"], sample_dists):   
    plt.plot(timeStep,dataframe[plant], color="black", label="Measurement")
    
    plt.plot(x, sample_dist.cdf(x), color="green", label="Idealised")
    delay = length*myclip_a
    
    t10 = sample_dist.ppf(0.1)-delay
    t50 = sample_dist.ppf(0.5)-delay
    t90 = sample_dist.ppf(0.9)-delay
    #print([t10,t50,t90])

    t10= t10/0.535812
    t50= t50/1.678347
    t90= t90/3.889720
    tau_105090 =(t10+t50+t90)/3
    
    tau = sample_dist.std() # check why this is true
    
    print(f"Standard deviation of arrival time: {tau}s, Zeit-Prozentkennlinien-Verfahren {tau_105090}s")

    x = numpy.linspace(delay,250,1000)
    U = 1-numpy.exp(-(x-delay)/tau) 
    U = U - (x-delay)/tau *   numpy.exp(-(x-delay)/tau)

    plt.plot(x,U, color="red", label="Approximation with PT2 lag")
    plt.legend()
    plt.xlim([10,130])
    plt.show()
```
